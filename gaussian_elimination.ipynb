```python
%autoreload 2
from src import eliminacion_gaussiana

Ab = [[1, 2, 0], [1, -1, 0]]

eliminacion_gaussiana(Ab)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Obtener los coeficientes de las ecuaciones a partir de Ab
x1, y1, c1 = Ab[0]  # Coeficientes de la primera ecuación
x2, y2, c2 = Ab[1]  # Coeficientes de la segunda ecuación

# Generar valores para graficar
x_values = np.linspace(-3, 3, 100)
# Ecuaciones de las rectas
y1_values = (c1 - x1 * x_values) / y1
y2_values = (c2 - x2 * x_values) / y2

# Graficar
plt.figure(figsize=(10, 6))
plt.plot(x_values, y1_values, label=f'Ecuación 1: {x1}x + {y1}y = {c1}', color='blue')
plt.plot(x_values, y2_values, label=f'Ecuación 2: {x2}x + {y2}y = {c2}', color='orange')

# Graficar el punto de intersección
Ab_np = np.array(Ab)
interseccion = np.linalg.solve(Ab_np[:, :-1], Ab_np[:, -1])
plt.plot(interseccion[0], interseccion[1], 'ro', label='Punto de intersección')

# Configuraciones de la gráfica
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')
plt.title('Gráfica de Ecuaciones y Punto de Intersección')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.show()  # Muestra la gráfica
```
